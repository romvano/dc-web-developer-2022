# Введение в алгоритмы и структуры данных

Опустимся на уровень языков чуть менее верхнеуровневых, чем Python, в которых на уровне языка не реализованы конструкции типа итератора, которые замещают собой функционал указателей. Приведем несколько определений.

- **Указатель** - это объект, в котором хранится адрес (по сути - номер) ячейки оперативной памяти, где записаны интересующие нас данные.

- **Ссылка** - это объект, указывающий на определенную область памяти.

Главное отличие ссылки от указателей: указатель - это целое число со всеми вытекающими из этого свойствами, в то время как для ссылки доступны только 2 операции: копирование и разыменование в указатель.

В питоне переменные - это ссылки: они не являются адресами памяти, но по сути являются объектами, которые ссылается на нужные данные. Кроме того, ссылками являются и объекты, хранящиеся в коллекциях. Эти два факта позволяют программировать на питоне, вообще не думая о структурах данных и алгоритмах, которые в нём заложены. 

### Зачем нужны структуры данных?

Базовые структуры данных лежат в основе любой коллекции данных. По сути академическая наука о структурах данных - это о том, как, храня в оперативной памяти минимум информации, понимать, как получить определенные данные. В качестве примера рассмотрим язык C. У него, в отличие от питона, каждый элемент массива должен быть одного и того же типа. Это позволяет хранить в памяти только адрес начала данных, а остальные данные можно получить, отступив от него `N * X` байт, где `X` - это количество байт, занимаемых единственным элементом этого типа, а N - номер элемента, который мы хотим получить из массива.



## Структура данных

Питоновские коллекции - это подмножество структур данных. Приведем определение:

**Структура данных** - это программная единица, в которой хранятся однотипные или логически связанные между собой данные. Структура данных чаще всего (но не всегда) должна поддерживать следующие операции:

- добавление данных
- поиск данных
- изменение данных
- удаление данных

Примеры структур данных: список, массив, очередь, словарь, множество и т.д...

## Алгоритм

**Алгоритмом** будем называть формально описанную вычислительную процедуру, которая получает входные данные и выдает некоторый результат (также в виде данных). Определение очень похоже на описание того, для чего используются функции. Таким образом, получается, что функция - это способ описания алгоритма в таком виде, чтобы компьютер мог этот алгоритм исполнить. Часто алгоритмы определены над конкретной структурой данных.

Одни и те же действия всегда можно сделать более эффективно или менее эффективно. Под **эффективностью** в теории алгоритмов имеются в виду 2 характеристики:

- количество потребляемой алгоритмом дополнительной памяти
- время работы алгоритма
- и менее часто рассматриваемые характеристики: количество обращений к жесткому диску, количеством операций сравнения данных и т.д.

### Оценка эффективности алгоритмов

В теории алгоритмов оценка эффективности рассматривает не конкретное время или память выполнения алгоритма, поскольку один и тот же код может быть запущен в системах с различными характеристиками производительности.

Введем обозначения:

- `M(n)` - ассимптотическая оценка объема дополнительной памяти, которую использует алгоритм, в зависимости от размера входных данных (от количества элементов входной структуры данных)
- `T(n)` - ассимптотическая оценка времени работы алгоритма в зависимости от размера входных данных

### Асимптотические оценки

**Асимптота** в математике - это линия в системе координат, к которой график функции стремится, но никогда не касается. В теории алгоритмов в краевом случае график оценки времени или памяти алгоритма будет совпадать с ассимптотической оценкой.

Как для времени, так и для памяти, используемой алгоритмом, вводится 3 вида оценок:


#### Тета-оценка



- при достаточно больших n оценка сложности алгоритма находится между двух графиков из одного семейства функций

$$ f(n) \in \theta(g(n)) = { \exists  c_1 > 0, c_2 > 0, n_0 > 0 ∶ 0 ≤ c_1 g(n) ≤ f(n) ≤ c_2 g(n), \forall n > n_0} $$


#### Оценки Омега-большое и омега-малое



- при достаточно больших n оценка сложности алгоритма всегда выше ($\omega$) / всегда выше или совпадает ($\Omega$) с некоторым графиком `g(n)`. Таким образом, мы можем считать `g(n)` нижней асимптотической границей сложности рассматриваемого алгоритма `f(n)`

$$ f(n) \in \omega(g(n)) = { \exists c > 0, n_0 > 0 : 0 < cg(n) < f(n), \forall n > n_0 } $$
$$ f(n) \in \Omega(g(n)) = { \exists c > 0, n_0 > 0 : 0 < cg(n) <= f(n), \forall n > n_0 } $$


#### Оценки О-большое и о-малое



- при достаточно больших n оценка сложности алгоритма всегда ниже (`o`) / всегда ниже или совпадает (`O`) с некоторым графиком `g(n)`. Таким образом, мы можем считать `g(n)` нижней асимптотической границей сложности рассматриваемого алгоритма `f(n)`

$$ f(n) \in o(g(n)) = { \exists c > 0, n_0 > 0 : 0 < f(n) < cg(n), \forall n > n_0 } $$
$$ f(n) \in O(g(n)) = { \exists c > 0, n_0 > 0 : 0 < f(n) <= cg(n), \forall n > n_0 } $$

Чаще всего пользуются оценкой `O(n)`.

### Оценка сложности поиска по списку

Поиск по списку буквально проводится по следующему алгоритму: идем из начала в конец, пока не встретим требуемый элемент. Время прохода по списку линейно зависит от длины списка, поэтому сложность такого алгоритма - **O(N)**

```python
lst = [5, 2, 6, 3, 8]
goal = 9

flag = False
for element in lst:
    if element == goal:
        flag = True
        break
        
print("Элемент найден" if flag else "Элемент не найден")
```

## Оценка сложности сортировки списка

Рассмотрим самый простой алгоритм сортировки списков - сортировку "пузырьком".

```python
# вспомогательная функция, которая меняет местами элементы списка с индексами pos1 и pos2
def _swap(lst: list, pos1: int, pos2: int):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp

    
# функция, которая реализует пузырьковую сортировку внутри уже определенного списка
def bubble_sort(lst: list):
    for i in range(len(lst)):            # первый проход по всему списку - n итераций, т.е. O(n)
        for j in range(i, len(lst)):     # вложенный проход по всему списку - n-i итераций, т.е. O(n)
            if lst[i] > lst[j]:          # i всегда < j. Если левый элемент больше правого, меняем их местами
                _swap(lst, i, j)
```

```python
lst = [5, 3, 6, 3, 2, 9]
bubble_sort(lst)
lst
```

Общая эффективность алгоритма:
- память: `O(1)`, поскольку дополнительная память не используется
- время: `O(n) * O(n) = O(n^2)`, поскольку есть двойной вложенный цикл, каждый внутренний цикл которого зависит от длины списка

А теперь рассмотрим сортировку, которую эффективно использовать для списков, в которых много повторяющихся элементов, - сортировку повторений:

```python
def repeatition_sort(lst: list):
    l = max(lst) + 1                         # первый проход по длине lst: поиск максимального элемента - O(n)
    counts = [0] * l                         # создаем новый список = дополнительная память O(k), где k=max(lst)
    for item in lst:
        counts[item] += 1                    # подсчитываем количество нахождений элементов в исходном список,
                                             # записывая их в новый список на позицию, номер которой равен
                                             # значению элемента. Это еще +O(n)
    i = 0    
    for count_i, count in enumerate(counts): # дальше записываем в исходный список элементы по столько подряд
        for _ in range(count):               # раз, сколько записали во второй список. Это еще O(n) + O(k)
            lst[i] = count_i
            i += 1
```

```python
lst = [4, 4, 4, 4, 4, 5, 5, 2, 2, 2, 4, 1, 2, 2, 2, 5, 2, 2, 5, 3, 7, 0]
repeatition_sort(lst)
lst
```

Общая эффективность алгоритма:

- память: `O(k)`, где `k = max(lst)`
- время: `O(n) + O(k) + O(n) + O(n) + O(k) = O(n) + O(k) = O(max(n, k))`

## Связный список

Массив и список - это структуры данных, которые позволяют быстро получать доступ к элементу по индексу. Но при этом вставка/удаление в начало или в середину новых элементов - это дорогостоящая по времени операция. В противоположность этому есть абстрактная структура данных **связный список**, в котором нету индексов, а связи между элементами прописываются внутри самих элементов.

```python
# в случае реализации абстрактных структур данных нам понадобится
# специальная обертка над каждым элементом структуры

class Node:
    def __init__(self, data=None, next_node=None):
        self.data = data        # в каждом узле списка храним некоторые данные (любого типа)
        self.next = next_node   # и указатель на следующий элемент


# а сам список будет оперировать уже этими обертками
class LinkedList:
    def __init__(self):
        self.head = None  # по умолчанию в списке нет элементов -> указатель на начальный элемент отсутствует

    def is_empty(self):
        """Метод для проверки, пустой ли список. Сложность по времени О(1)."""
        return self.head is None

    def append(self, data):
        """Добавление элемента в конец списка. Сложность по времени О(N)."""
        new_node = Node(data)       # создаем новый элемент списка
        if self.head is None:       # если список пустой, то указываем на новый начальный элемент
            self.head = new_node
            return
        last_node = self.head       # если список не пустой, то проходимся по каждому элементу,
        while last_node.next:       # пока не найдем конец списка
            last_node = last_node.next
        last_node.next = new_node   # когда находим элемент, у которого нет следующего элемента, то он и есть
                                    # последний. Значит, к нему и нужно прилепить новый элемент

    def prepend(self, data):
        """Добавление элемента в начало списка. Сложность по времени О(1)."""
        new_node = Node(data)       # создаем новый элемент списка
        new_node.next = self.head   # следующим для него будет начальный элемент списка
        self.head = new_node        # и подменяем начальный элемент на новый

    def delete(self, value):
        """Удаление элемента по значению. Сложность по времени О(N)."""
        if self.is_empty():                                            # 1. если список пуст, то в нем нет
            raise ValueError(f"No such element in list: {value}.")     #    удаляемого элемента
            
        if self.head.data == value:                                    # 2. если удаляемый элемент - головной,
            self.head = self.head.next                                      #    то просто подменим начало списка
            return
        
        current_node = self.head                                       # 3. а дальше проходимся по списку,
        prev_node = None                                               #    каждый раз держа в памяти текущий
        while current_node is not None and current_node.data != value: #    и предыдущий элементы, пока не
            prev_node = current_node                                   #    дойдем до конца или не найдем нужный
            current_node = current_node.next                           #    элемент
            
        if current_node is None:                                       # 4. если мы дошли до конца списка, то
            raise ValueError(f"No such element in list: {value}.")     #    искомого значения в нем нет
            
        prev_node.next = current_node.next                             # 5. а если нашли значение, то подменяем
                                                                       #    next предыдущего элемента на следую-
                                                                       #    щий от найденного
    def __str__(self):
        """Вывод списка на экран. Сложность по времени О(N)."""
        result_str = ""
        current_node = self.head
        while current_node:
            result_str += f" {current_node.data}"
            current_node = current_node.next
        return result_str
```

### Пример использования

```python
linked_list = LinkedList()
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)
print(linked_list)      # Вывод: 1 2 3

linked_list.prepend(0)
print(linked_list)      # Вывод: 0 1 2 3

linked_list.delete(2)
print(linked_list)      # Вывод: 0 1 3

linked_list.delete(0)
print(linked_list)      # Вывод: 1 3

linked_list.delete(25)  # Ошибка
```

### Более реальный пример

Допустим, мы проводим большой онлайн-экзамен, и пообещали, что результаты будем публиковать по мере проверки в алфавитном порядке. В этом случае нам нужно хранить отсортированный список студентов, из которого мы хотим эффективно удалять наименьший (лексикографически) элемент. Связный список как раз позволяет быстро удалять элементы из начала. Чтобы всегда в первую очередь проверять студента, который идет первым по алфавиту, нам нужно добавлять новые элементы в список сразу с учетом фамилии, т.е. элементы в нашем списке всегда будут отсортированы.

Реализуем структуру, в которой:
- удаление элементов будет занимать О(1) времени
- добавление элемента будет занимать О(N) времени (добавление будет происходить в середину списка по правилу)

```python
from dataclasses import dataclass
import datetime

@dataclass
class Student:
    surname: str
    name: str
    pathronymic: str
    birth_date: datetime.date
    exam: str
    next_student: "Student" = None
        
    @property
    def fio(self):
        return " ".join([self.surname.upper(), self.name.upper(), self.pathronymic.upper()])
    
    def __str__(self):
        return self.fio
    
    def __repr__(self):
        return self.fio
        
        
class StudentList:
    def __init__(self):
        self.head = None
        
    def insert(self, student: Student):
        if self.head == None:
            self.head = student
            return
        if student.fio < self.head.fio:
            student.next_student = self.head
            self.head = student
            return
        current_node = self.head
        prev_node = None
        while current_node is not None and current_node.fio < student.fio:
            prev_node = current_node
            current_node = current_node.next_student
        prev_node.next_student = student
        student.next_student = current_node
        
    def pop(self):
        first_student = self.head
        self.head = self.head.next_student
        return first_student
```

```python
student_a = Student("Абрамов", "Абрам", "Абрамович", datetime.date(2000, 1, 1), "")
student_b = Student("Большой", "Борис", "Борисович", datetime.date(1998, 2, 2), "")
student_z = Student("Яшкевич", "Ярополк", "Янович", datetime.date(1389, 3, 3), "")

student_list = StudentList()
student_list.insert(student_b)
student_list.insert(student_z)
student_list.insert(student_a)

student_list.pop()
```

# Очередь и стек

**Очередь** и **стек** - это еще более абстрактные структуры данных, в которых можно содержать любые объекты. Их можно реализовать и через питоноский список, и через связный список.

## Очередь

**Очередь (queue)** - это структура данных, которая работает по принципу FIFO (First-In-First-Out), что означает, что элементы добавляются в конец очереди, а извлекаются из начала очереди, подобно очереди в реальной жизни, например, в магазине: первый клиент пришел - первый ушел.

Очередь имеет две основные операции:

- **Enqueue** (добавление): добавление элемента в конец очереди.
- **Dequeue** (удаление): извлечение элемента из начала очереди.



```python
class Queue:
    def __init__(self):
        """В питоне, естественно, проще реализовать очередь поверх встроенного списка."""
        self._queue = []

    def is_empty(self):
        """Проверяет, является ли очередь пустой."""
        return len(self._queue) == 0

    def enqueue(self, item):
        """Добавление элемента всегда происходит в конец очереди. Сложность при реализации на списке - O(1)."""
        self._queue.append(item)

    def dequeue(self):
        """Удаление - всегда первого элемента. Поскольку элемент всегда первый - сложность тоже О(1)."""
        return self._queue.pop(0)  # свалим вызов исключений при пустом списке на сам питоновский список

    def peek(self):
        """Метод, который возвращает первый элемент без удаления."""
        return self._queue[0]

    def size(self):
        """Возвращает размер очереди."""
        return len(self._queue)
```

### Пример использования очереди


```python
my_queue = Queue()

my_queue.enqueue(1)
my_queue.enqueue(2)
my_queue.enqueue(3)

print(my_queue.dequeue())  # Вывод: 1
print(my_queue.peek())     # Вывод: 2
print(my_queue.size())     # Вывод: 2
```

## Стек

**Стек (stack)** - это абстрактная структура данных, которая работает по принципу LIFO (Last-In, First-Out) или "последний вошел, первый вышел". Это означает, что элементы добавляются и удаляются только с одного конца структуры данных, который обычно называется "вершиной" (или "top").

Основные операции со стеком:

- **Push** (добавление): новый элемент помещается на вершину стека.
- **Pop** (удаление): элемент удаляется с вершины стека.
- **Peek** (просмотр): получение значения элемента на вершине стека без его удаления.



Со стеком вы уже сталкивались, например, когда смотрели на сообщения об исключениях. Функции в программировании при вызове складываются как раз в стек, чтобы при отработке одной функции можно было вернуться в предыдущую.

```python
class Stack:
    def __init__(self):
        """Стек также можно реализовать как на питоновском списке, так и на связном списке."""
        self._items = []

    def is_empty(self):
        """Проверяет, пуст ли стек."""
        return len(self._items) == 0

    def push(self, item):
        """Добавляет элемент в конец списка."""
        self._items.append(item)

    def pop(self):
        """Возвращает и удаляет последний элемент из стека (элемент-вершину стека)."""
        if not self.is_empty():
            return self._items.pop()

    def peek(self):
        """Возвращает вершину стека."""
        if not self.is_empty():
            return self._items[-1]

    def size(self):
        """Возвращает размер стека."""
        return len(self._items)
```

### Пример использования стека

```python
stack = Stack()

stack.push(1)
stack.push(2)
stack.push(3)

print("peek:", stack.peek())            # Вывод: 3
print("pop:", stack.pop())              # Вывод: 3
print("size after pop:", stack.size())  # Вывод: 2
```

### Простой парсер математических выражений

Опишем простейший парсер математических выражений. Для этого нам нужно будет хранить два стека: операндов и операций. Не будем усложнять этот парсер приоритетом операций.

```python
from queue import LifoQueue


def parse_expression(expression):
    operators = LifoQueue()  # заведем стек для операторов
    operands = LifoQueue()   # и еще один для операндов
    
    expression = expression.replace("(", "").replace(")", "")  # скобки нам не нужны, все операции - одного приоритета
    for element in expression.split()[::-1]:
        if element in {"+", "-"}:        # идем с начала строки и операторы кладем в один стек
            operators.put(element)
        else:
            operands.put(float(element)) # а операнды - в другой
            
    while operands.qsize() > 1:
        operator = operators.get()       # пока в операндах не останется только один элемент,
        operand1 = operands.get()        # берем два последних операнда и один последний оператор и проводим
        operand2 = operands.get()        # вычисление на основе этого. Результат кладем снова в операнды
        
        if operator == "+":
            operands.put(operand1 + operand2)
        elif operator == "-":
            operands.put(operand1 - operand2)
    
    return operands.get()                # последний оставшийся операнд - ответ


ex = "5 + 3 - 2 - (-7)"
parse_expression(ex)
```
