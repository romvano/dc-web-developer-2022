```python
#  ,        pos1  pos2
def _swap(lst: list, pos1: int, pos2: int):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp

    
# ,        
def bubble_sort(lst: list):
    for i in range(len(lst)):            #      - n , .. O(n)
        for j in range(i, len(lst)):     #      - n-i , .. O(n)
            if lst[i] > lst[j]:          # i  < j.     ,   
                _swap(lst, i, j)
```

```python
lst = [5, 3, 6, 3, 2, 9]
bubble_sort(lst)
lst
```

```python
def repeatition_sort(lst: list):
    l = max(lst) + 1                         #     lst:    - O(n)
    counts = [0] * l                         #    =   O(k),  k=max(lst)
    for item in lst:
        counts[item] += 1                    #       ,
                                             #       ,   
                                             #  .   +O(n)
    i = 0    
    for count_i, count in enumerate(counts): #         
        for _ in range(count):               # ,     .   O(n) + O(k)
            lst[i] = count_i
            i += 1
```

```python
lst = [4, 4, 4, 4, 4, 5, 5, 2, 2, 2, 4, 1, 2, 2, 2, 5, 2, 2, 5, 3, 7, 0]
repeatition_sort(lst)
lst
```

```python
#        
#      

class Node:
    def __init__(self, data=None, next_node=None):
        self.data = data        #        ( )
        self.next = next_node   #     


#        
class LinkedList:
    def __init__(self):
        self.head = None  #       ->     

    def is_empty(self):
        """  ,   .    (1)."""
        return self.head is None

    def append(self, data):
        """    .    (N)."""
        new_node = Node(data)       #    
        if self.head is None:       #   ,      
            self.head = new_node
            return
        last_node = self.head       #    ,     ,
        while last_node.next:       #     
            last_node = last_node.next
        last_node.next = new_node   #   ,     ,    
                                    # . ,       

    def prepend(self, data):
        """    .    (1)."""
        new_node = Node(data)       #    
        new_node.next = self.head   #       
        self.head = new_node        #      

    def delete(self, value):
        """   .    (N)."""
        if self.is_empty():                                            # 1.   ,    
            raise ValueError(f"No such element in list: {value}.")     #     
            
        if self.head.data == value:                                    # 2.    - ,
            self.head = self.head.next                                      #        
            return
        
        current_node = self.head                                       # 3.     ,
        prev_node = None                                               #         
        while current_node is not None and current_node.data != value: #      ,  
            prev_node = current_node                                   #          
            current_node = current_node.next                           #    
            
        if current_node is None:                                       # 4.      , 
            raise ValueError(f"No such element in list: {value}.")     #        
            
        prev_node.next = current_node.next                             # 5.    ,  
                                                                       #    next    -
                                                                       #      
    def __str__(self):
        """   .    (N)."""
        result_str = ""
        current_node = self.head
        while current_node:
            result_str += f" {current_node.data}"
            current_node = current_node.next
        return result_str
```

```python
linked_list = LinkedList()
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)
print(linked_list)      # : 1 2 3

linked_list.prepend(0)
print(linked_list)      # : 0 1 2 3

linked_list.delete(2)
print(linked_list)      # : 0 1 3

linked_list.delete(0)
print(linked_list)      # : 1 3

linked_list.delete(25)  # 
```

```python
from dataclasses import dataclass
import datetime

@dataclass
class Student:
    surname: str
    name: str
    pathronymic: str
    birth_date: datetime.date
    exam: str
    next_student: "Student" = None
        
    @property
    def fio(self):
        return " ".join([self.surname.upper(), self.name.upper(), self.pathronymic.upper()])
    
    def __str__(self):
        return self.fio
    
    def __repr__(self):
        return self.fio
        
        
class StudentList:
    def __init__(self):
        self.head = None
        
    def insert(self, student: Student):
        if self.head == None:
            self.head = student
            return
        if student.fio < self.head.fio:
            student.next_student = self.head
            self.head = student
            return
        current_node = self.head
        prev_node = None
        while current_node is not None and current_node.fio < student.fio:
            prev_node = current_node
            current_node = current_node.next_student
        prev_node.next_student = student
        student.next_student = current_node
        
    def pop(self):
        first_student = self.head
        self.head = self.head.next_student
        return first_student
```

```python
class Queue:
    def __init__(self):
        """ , ,      ."""
        self._queue = []

    def is_empty(self):
        """,    ."""
        return len(self._queue) == 0

    def enqueue(self, item):
        """      .      - O(1)."""
        self._queue.append(item)

    def dequeue(self):
        """ -   .     -   (1)."""
        return self._queue.pop(0)  #          

    def peek(self):
        """,      ."""
        return self._queue[0]

    def size(self):
        """  ."""
        return len(self._queue)
```

** (stack)** -    ,     LIFO (Last-In, First-Out)  " ,  ".  ,           ,    "" ( "top").

   :

- **Push** ():      .
- **Pop** ():     .
- **Peek** ():         .

```python
class Stack:
    def __init__(self):
        """       ,     ."""
        self._items = []

    def is_empty(self):
        """,   ."""
        return len(self._items) == 0

    def push(self, item):
        """    ."""
        self._items.append(item)

    def pop(self):
        """       (- )."""
        if not self.is_empty():
            return self._items.pop()

    def peek(self):
        """  ."""
        if not self.is_empty():
            return self._items[-1]

    def size(self):
        """  ."""
        return len(self._items)
```

###   

```python
stack = Stack()

stack.push(1)
stack.push(2)
stack.push(3)

print("peek:", stack.peek())            # : 3
print("pop:", stack.pop())              # : 3
print("size after pop:", stack.size())  # : 2
```

```python
from queue import LifoQueue


def parse_expression(expression):
    operators = LifoQueue()  #    
    operands = LifoQueue()   #     
    
    expression = expression.replace("(", "").replace(")", "")  #    ,   -  
    for element in expression.split()[::-1]:
        if element in {"+", "-"}:        #          
            operators.put(element)
        else:
            operands.put(float(element)) #   -  
            
    while operands.qsize() > 1:
        operator = operators.get()       #        ,
        operand1 = operands.get()        #          
        operand2 = operands.get()        #    .     
        
        if operator == "+":
            operands.put(operand1 + operand2)
        elif operator == "-":
            operands.put(operand1 - operand2)
    
    return operands.get()                #    - 


ex = "5 + 3 - 2 - (-7)"
parse_expression(ex)
```
